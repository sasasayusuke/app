```python
# モジュールをインポート

from wordpress_xmlrpc import Client, WordPressPost, methods

def postArticle(title, content, category): 

    # アクセス情報の設定
    AUTH_USER = 'takashi'
    AUTH_PASS = 'wp!YTss1123'

    # 投稿するURLの設定
    END_POINT_URL ="https://vipperlife.com/xmlrpc.php"
    # END_POINT_URL ="https://aiillustlife.online/xmlrpc.php"

    client = Client(END_POINT_URL, AUTH_USER, AUTH_PASS)

    # 投稿内容# コンテンツを作成する。
    post = WordPressPost()
    post.title = title
    post.content = content
    post.terms_names = {"post_tag": [category, "WordPress"], "category": [category]}

    # コンテンツを投稿する。
    post_id = client.call(methods.posts.NewPost(post))
    return post_id

postArticle(21, 21, "test")
```

```python
import requests
from bs4 import BeautifulSoup

def getTopNewsLink(category="政治"): 
    # リクエストを送信してHTMLコードを取得する
    url = f"https://www.2nn.jp/{getCategory(category)}/"
    response = requests.get(url)
    html = response.text

    # BeautifulSoupを使ってHTMLコードからリンクを抽出する
    soup = BeautifulSoup(html, "html.parser")
    return soup.select("#top-news h1 a")[0]["href"]

getTopNewsLink()
```
